```python
from random import randrange
from datetime import datetime

# 은행 계좌 초기 셋팅

# 예금주, 초기 잔액 name, first_money
class Account:
  # 클래스 변수
  bank = "SC은행"

  def __init__(self, name, first_money):
    self.name = name # 예금주
    self.money = first_money # 잔고
    self.account_number = f"{randrange(100, 1000)}-{randrange(10, 100)}-{randrange(100000, 1000000)}" # 계좌 번호, 초기 생성시 랜덤 3-2-6 형태로 생성
    self.deposit_count = 0 # 입금 카운트
    self.withdraw_count = 0 # 출금 카운트 대출 전용
    self.deposit_list = [] # 입금 내역 리스트
    self.withdraw_list = [] # 출금 내역 리스트
    self.debt = 0 # 빚

  # 입금 메서드
  def deposit(self, deposit_money):
    if deposit_money >= 1:
      now = datetime.now() # 현재 시간을 가져오는 함수
      self.money += deposit_money
      self.deposit_count += 1
      self.deposit_list.append(now.strftime('%Y-%m-%d %H:%M:%S')) # 년,달,날,시간,분,초로 출력

    # 입금 카운터가 5일 경우 이자 지급 메서드 실행
    if self.deposit_count == 5:
      self.interest_payments()

    return format(self.money, ',')

  # 출금 메서드
  def withdraw(self, withdraw_money):
    if withdraw_money < self.money:
      now = datetime.now()
      self.money -= withdraw_money
      self.withdraw_list.append(now.strftime('%Y-%m-%d %H:%M:%S'))

    # 남은 빚이 있을 경우 빚 이자 메서드 실행
    if self.debt > 0:
      self.deposit_count += 1
      if self.deposit_count == 5:
        self.debt_payments()

    return format(self.money, ',')

  # 정보 출력 메서드
  # 은행이름, 예금주, 계좌번호, 잔액 출력
  def display_info(self):
    user = {"은행이름" : self.bank,
            "예금주" : self.name,
            "계좌번호" : self.account_number,
            "잔액" : format(self.money, ','),
            "남은 빚" : format(self.debt, ',')
            }

    return user

  # 이자 지급 메서드 5번마다 1% 이자가 잔고에 추가.
  def interest_payments(self):
      self.money += int(self.money * 0.01)
      self.deposit_count = 0

  # 입금 내역 출력 메서드
  def deposit_history(self):
    return self.deposit_list

  # 출금 내역 출력 메서드
  def withdraw_history(self):
    return self.withdraw_list

  # 입출금 내역 동시 출력 메서드
  def all_history(self):
    return self.deposit_list, self.withdraw_list


  # 루브릭 3 추가 메서드.
  # 출금시 잔액이 -이면 대출메서드 실행.
  def loan_money(self, debt_money):
    self.debt = debt_money
    self.money += debt_money

  # 남은 빚 확인 메서드
  def debts(self):
    return format(self.debt, ',')

  # 빚 이자 메서드
  def debt_payments(self):
    self.debt += int(self.debt * 0.01)
    self.deposit_count = 0

  # 빚 갚기 메서드
  def payback(self, payback_money):
    if self.debt >= payback_money:
      self.debt -= payback_money

    return self.debt
```

```python
# 메인 코드
if __name__ == "__main__":
  name = input("예금주 이름을 입력하세요.")
  first_money = int(input("초기 잔액을 입력하세요."))

  newaccount = Account(name, first_money)

  while(1):
    choice = int(input("1: 현재 고객 정보 확인하기 2: 입금하기 3: 출금하기 4: 입금 내역 출력하기 5: 출금 내역 출력하기 6: 입출금 내역 출력하기 7: 대출하기 8: 빚 갚기 9 : 남은 빚 확인 0: 종료하기 "))

    if choice == 0:
      break

    elif choice == 1:
      user = newaccount.display_info()
      print(user)

    # 입금하기
    elif choice == 2:
      deposit_money = int(input("얼마를 입금하시겠습니까? "))
      now_money = newaccount.deposit(deposit_money)
      print(f"현재 남은 잔고는 {now_money}원 입니다.")

    # 출금하기
    elif choice == 3:
      withdraw_money = int(input("얼마를 출금하시겠습니까? "))
      now_money = newaccount.withdraw(withdraw_money)
      print(f"현재 남은 잔고는 {now_money}원 입니다.")

    elif choice == 4:
      deposit_list = newaccount.deposit_history()
      print(deposit_list)

    elif choice == 5:
      withdraw_list = newaccount.withdraw_history()
      print(withdraw_list)

    elif choice == 6:
      deposit_list, withdraw_list = newaccount.all_history()
      print(f"입금 내역 : {deposit_list}")
      print(f"출금 내역 : {withdraw_list}")

    elif choice == 7:
      debt_money = int(input("얼마를 대출받으시겠습니까?"))
      newaccount.loan_money(debt_money)

    elif choice == 8:
      payback_money = int(input("얼마를 갚으시겠습니까?"))
      newaccount.payback(payback_money)

    elif choice == 9:
      debt = newaccount.debts()
      print(f"남은 빚은 {debt}원 입니다.")

    else :
      print("잘못 입력했습니다. 다시 입력해주세요")
```
